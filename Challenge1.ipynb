```python
# Write a program to work out how many days you have lived for.
import datetime
now = datetime.date.today()
string_date = input("Input your birthday (dd/mm/yyyy): ")
birthday = datetime.date(int(string_date[6:10]), int(string_date[3:5]), int(string_date[:2]))
time_passed = now - birthday
print (str(time_passed.days) + " days")      #.days element is the number of days in the date object

```

```python
# Write a program to work out how many seconds you have lived for.
print (str(time_passed.total_seconds()) + " seconds")          #assuming the person is born at midnight
```

```python
# Write a program that will accept a date of birth and determine whether one 
# is considered an adult (i.e. 21 years old).
import datetime
now = datetime.date.today()
string_date = input("Input your birthday (dd/mm/yyyy): ")
adult_date = datetime.date(int(string_date[6:10]) + 21, int(string_date[3:5]), int(string_date[:2]))  #day when the person becomes adult
if adult_date <= now:     #adult date is today or has been passed
    print ("Adult")
else:
    print ("Not Adult")
```

```python
# Write a program that will give the answer to logic gate questions e.g.
# Enter logic gate : OR
# Enter first input : 1
# Enter second input :0
# Result = 1
# It should work for the logic gates OR, XOR, NAND and NOR gates.

true_false = [False, True]      #index 1 is true, index 0 is false
logic_gate = input("Enter logic gate: ").upper()
first_input = input("Enter first input: ")
first_input = true_false[int(first_input)]
second_input = input("Enter second input: ")
second_input = true_false[int(second_input)]

if logic_gate == "OR":
    print (true_false.index(first_input or second_input))
elif logic_gate == "XOR":
    print (true_false.index(first_input != second_input))      #odd number of inputs, if they are equal, then its true
elif logic_gate == "NAND":
    print (true_false.index(not first_input and second_input))
elif logic_gate == "NOR":
    print (true_false.index(not first_input or second_input))
else:
    print ("Invalid logic gate")
    
```

```python
# Write a program that will display all the factors of an input number that
# are bigger than 1.
# (e.g. the factors of the number 12 are 6,4,3 and 2 because they divide
# into 12 exactly).
# Hint: To find out whether a number X is a factor of Y use:
# If Y mod X == 0 (there is nothing remaining when Y is divided by X)
number = int(input("Input number: "))
factors = []
for i in range(2, int(number ** (1/2) + 1)):             #all factors of the number (or at least their mirror) would appear 
    if number % i == 0:        #i is a factor of number
        factors.append(i)
print ("The factors of the number are: ", end = '')
for i in factors:
    print (i, end = ', ')
start = 1          #start to print from the back when finding the "mirror" image of the factors
if number // factors[-1] == factors[-1]:     #number is a perfect square, don't repeat the square factor (e.g. 10 for 100)
    start = 2
for i in range(start, len(factors)):           #everything except last element
    print (number // factors[-i], end = ', ')
print (number // factors[0])
```
