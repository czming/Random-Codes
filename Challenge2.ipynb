```python
''' 
Write a program for a game where the computer generates a
random starting number between 20 and 30.
The player and the computer can remove 1,2 or 3 from the number
in turns. Something like this:

Starting number: 25
How many do you want to remove? 3
22 left
Computer removes 2
20 left
...

The player who has to remove the last value to bring the number
down to 0 is the loser.

1 left
Computer removes 1
You win!

Hard: Get the computer to employ a strategy to try and win. 
'''
from random import randint as r
#rough strategy implemented (probably not the best but it depends on which number is starting anyway)
#another strat is to model out all the possible cases and see those that will result in victory and see which step the computer should take (like see the next step and see if whatever the user does, the movements, in an array, are still in those that will lead to victor)
number = r(20,30)                #random value between 20 and 30 (inclusive)
print ("Starting number: " + str(number))
while True:             #what happens if less than 0 in the rules?
    user_remove = 0
    while user_remove < 1 or user_remove > 3:     #until user input a value between 1 and 3 (inclusive)
        user_remove = int(input("How many do you want to remove? "))
    if user_remove >= number:      #number will be 0 or negative after deducting user_remove
        print ("You lose!")
        break
    else:
        number -= user_remove
        print (str(number) + " left")
    if number <= 4 and number != 1:     #if number is 1, computer has lost already 
        computer_remove = number - 1    #eg if number 4, computer will remove 3, forcing user to choose 1 and lose
    elif number == 5: 
        computer_remove = 1      #computer has alerady lost in this case since no matter what it chooses, the next number will be 2-4
    elif number <= 7:                 #wants to avoid falling into the range of 1-4 cause that's where the next player can win, it needs to make sure its the first to 4, so if its odd and everyone keeps playing 1, it will win
        computer_remove = number % 5     #want to force the number into 5 so whatever the user puts, the computer can win at the next step (since will be 2-4 at the next step no matter what the user puts)
    else:
        computer_remove = 3    #fastest way to the bottom
    if computer_remove >= number:
        print ("Computer removes " + str(number))     #computer reduces nicely to 0
        print ("You win!")
        break
    else:
        number -= computer_remove
        print ("Computer removes " + str(computer_remove))
        print (str(number) + " left")
```

```python
"""###
Write a program for a Higher / Lower guessing game.
The computer randomly generates a sequence of up to 10 numbers
between 1 and 13. The player after seeing each number in turn
has to decide whether the next number is higher or lower.
If you get 10 guesses right you win the game.

Starting number : 12
Higher(H) or lower(L)? L
Next number 8
Higher(H) or lower(L)? L
Next number 11
You lose

Hints
- Use a condition controlled loop (do until, while etc) to control the
game.
- You do not need to remember all 10 numbers, but  just the current and
the next number.
- Remember to keep a count of the number of guesses. 
"""

from random import randint as r
number_array = []
for i in range(11):          #not necessary to store all the values, but I think it's more convenient to do so, need 11 numbers, because no need to guess first one
    curr_number = r(1,13)
    while len(number_array) > 0 and curr_number == number_array[i - 1]:       #in case of repeat values, len(number_array) > 0 added at the front to prevent IndexError
        curr_number = r(1,13)
    number_array.append(curr_number)
print (number_array)      #print out for checking purposes
print ("Starting number: " + str(number_array[0]))
curr = 0
correct = True
correct_guesses = 0
while curr < 10 and correct:
    correct = False
    user_choice = ""
    while user_choice != "H" and user_choice != "L": 
        user_choice = input("Higher(H) or lower(L): ").upper()
    if number_array[curr + 1] > number_array[curr]:
        correct = user_choice == "H"                 #if the user said higher and the next number is higher, user is correct
    else:
        correct = user_choice == "L"
    if correct: 
        correct_guesses += 1
    curr += 1
    print ("Next number: {0}".format(number_array[curr]))
if correct_guesses == 10:      #got all correct
    print ("You won the game with 10 correct guesses")
else:
    print ("You lost with {0} correct guesses".format(correct_guesses))
```

```python
'''
Extend the above Higher / Lower number guessing game by
- Giving the player two lives
- Allowing only H or L to be entered 
'''
from random import randint as r
number_array = []
for i in range(11):          #not necessary to store all the values, but I think it's more convenient to do so, need 11 numbers, because no need to guess first one
    curr_number = r(1,13)
    while len(number_array) > 0 and curr_number == number_array[i - 1]:       #in case of repeat values, len(number_array) > 0 added at the front to prevent IndexError
        curr_number = r(1,13)
    number_array.append(curr_number)
print (number_array)      #print out for checking purposes
print ("Starting number: " + str(number_array[0]))
curr = 0
correct = True
correct_guesses = 0
lives = 2       #user has two lives
while curr < len(number_array) - 1 and lives > 0:            #changed from above code due to possible changes in the array due to loss of lives
    correct = False
    user_choice = ""
    while user_choice != "H" and user_choice != "L": 
        user_choice = input("Higher(H) or lower(L): ").upper()
    if number_array[curr + 1] > number_array[curr]:
        correct = user_choice == "H"                 #if the user said higher and the next number is higher, user is correct
    else:
        correct = user_choice == "L"
    if correct: 
        correct_guesses += 1
    else:
        lives -= 1
        print("Incorrect, you have lost a life, lives left: {0}".format(lives))
        number_array.append(r(1,13))    #add another number at the end 
    curr += 1
    print ("Next number: {0}".format(number_array[curr]))      #even after dying, still show the next number so user knows what the next number is, if don't want to then add an if lives == 0 statement in the else statement above
if correct_guesses == 10:      #got all correct
    print ("You won the game with 10 correct guesses")
else:
    print ("All lives lost. You lost with {0} correct guesses".format(correct_guesses))
```

```python
'''
Write a program that will display a sentence backwards after being entered.
Note: You may not use the Python list reverse() function. ;)
'''
user_input = input("Enter sentence: ").split()
full_stop = False
reverse_input = []
for i in range(len(user_input) - 1, -1, -1):     
    if user_input[i][-1] == '.':                  #did not settle capital letter case though
        user_input[i] = user_input[i][:-1]     #remove the full stop
        full_stop = True
    reverse_input.append(user_input[i])
reverse_input = ' '.join(reverse_input)
print (reverse_input, end = '')
if full_stop:      #add back the full stop if present
    print (".", end = '')
```
